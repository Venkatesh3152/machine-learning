```python
import pandas as pd
```

```python
dataset=pd.read_csv("/content/data (1).csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dataset["Education_Level"] = le.fit_transform(dataset["Education_Level"])
```

```python
dataset["Gender"] = le.fit_transform(dataset["Gender"])
```

```python
dataset["Homeownership_Status"] = le.fit_transform(dataset["Homeownership_Status"])
```

```python
dataset["Gender"] = le.fit_transform(dataset["Gender"])
```

```python
dataset["Occupation"] = le.fit_transform(dataset["Occupation"])
```

```python
dataset["Location"] = le.fit_transform(dataset["Location"])
```

```python
dataset["Marital_Status"] = le.fit_transform(dataset["Marital_Status"])
```

```python
dataset["Employment_Status"] = le.fit_transform(dataset["Employment_Status"])
```

```python
dataset["Type_of_Housing"] = le.fit_transform(dataset["Type_of_Housing"])
```

```python

dataset["Primary_Mode_of_Transportation"] = le.fit_transform(dataset["Primary_Mode_of_Transportation"])
```

```python
dataset
```

```python

x = dataset[['Age','Education_Level','Occupation','Number_of_Dependents','Location','Work_Experience','Marital_Status','Employment_Status','Household_Size','Type_of_Housing','Primary_Mode_of_Transportation','Gender','Homeownership_Status']]
y = dataset['Income']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```
